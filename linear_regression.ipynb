Develop a Real world linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

Dataset: x- daily km

```python
x_train=np.array([1,2,3,4,5,7,10,12,15,20])
y_train=np.array([0,0.5,0.6,1,2,4,5,5.5,6,7])
plt.scatter(x_train,y_train)
```

estimate model para,m

```python
n=np.size(x_train)
n
```

no. of values=n=10

```python
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean,y_mean)
```

```python
sum_x=np.sum(x_train)
print('Sum of x =',sum_x)

sum_y=np.sum(y_train)
print('Sum of y =',sum_y)

sum_xy=np.sum(x_train*y_train)
print('Sum of x*y =',sum_xy)
```

calculate parameter of Beta1

```python
sum_xx=np.sum(x_train*x_train)
print('Sum of x^2 =',sum_xx)
```

```python
num1=n*sum_xy-(sum_x*sum_y)
den1=n*sum_xx-(sum_x**2)
beta_1=num1/den1

print('beta 1=',beta_1)
```

```python
beta_0=y_mean-(beta_1*x_mean)
beta_0
```

```python
ytrain_pred=beta_1*x_train+beta_0
print(ytrain_pred)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,ytrain_pred,'r')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
```

```python
from ast import Constant
xtrainnew=sm.add_constant(x_train) #adding constant
print(xtrainnew)
```

```python
ytrain_pred=sm.OLS(ytrain_pred,xtrainnew).fit() #training the model by importing OLS
```

```python
print(ytrain_pred.params)
```

```python
xtest=np.array([11,12,13,15,14,8,9,4,5,2])
xtestnew=sm.add_constant(xtest)
pred_y=ytrain_pred.predict(xtestnew)
print(pred_y)
```

```python
pred_y1=beta_1*xtest+beta_0
print(pred_y1)
#testing using equation
```

```python
ytest_real=y_train
```

```python
er = ytest_real - ytrain_pred.predict(xtrainnew) # Get predictions from the fitted model
print(er)
```

```python
er2=er**2
er2
```
