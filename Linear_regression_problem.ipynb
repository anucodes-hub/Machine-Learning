```python
import pandas as pd

# Create the DataFrame
data = {
    'Daily walking (kms)': [1, 2, 3, 4, 5, 7, 10, 12, 15, 20],
    'Weight loss (kgs)': [0, 0.5, 0.6, 1, 2, 4, 5, 5.5, 6, 7]
}
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
display(df)
```

define the features (X) and the target (y), and then split the dataset into training and testing sets.

'Daily walking (kms)' and 'Weight loss (kgs)'.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Daily walking (kms)'], df['Weight loss (kgs)'], color='green')
plt.xlabel('Daily walking (kms)')
plt.ylabel('Weight loss (kgs)')
plt.title('Daily Walking vs. Weight Loss')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df[['Daily walking (kms)']]
y = df['Weight loss (kgs)']

# Partition the dataset into 70% training and 30% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data size: {len(X_train)} samples")
print(f"Testing data size: {len(X_test)} samples")
```

build a Linear Regression model using the training data and make predictions on the test set.

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

print("Model training complete. Predictions made on test data.")
```

evaluate the performance of the model using common regression metrics such as R-squared, Mean Absolute Error (MAE), and Mean Squared Error (MSE).

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Evaluate the model
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Optionally, visualize the predictions vs actual values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual Weight Loss')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Weight Loss')
plt.xlabel('Daily Walking (kms)')
plt.ylabel('Weight Loss (kgs)')
plt.title('Linear Regression: Actual vs. Predicted Weight Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## Split Data into Training and Testing Sets

Define features (X) as 'Daily Walking' and target (y) as 'Weight Loss'. Then, split the dataset into 70% training and 30% testing data using `train_test_split` with `random_state=1`.


```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df[['Daily walking (kms)']]
y = df['Weight loss (kgs)']

# Partition the dataset into 70% training and 30% testing data with random_state=1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print(f"Training data size: {len(X_train)} samples (X_train)")
print(f"Testing data size: {len(X_test)} samples (X_test)")
print(f"Training target size: {len(y_train)} samples (y_train)")
print(f"Testing target size: {len(y_test)} samples (y_test)")
```

## Visualize Training Data


Generate a scatter plot to visualize the relationship between 'Daily Walking' and 'Weight Loss' specifically for the training data.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.xlabel('Daily walking (kms)')
plt.ylabel('Weight loss (kgs)')
plt.title('Training Data: Daily Walking vs. Weight Loss')
plt.grid(True)
plt.legend()
plt.show()
```

## Build and Train OLS Model


Use `statsmodels.api` to add a constant to the features and then train a Linear Regression (OLS) model using the training data. Print the regression parameters (coefficients).


```python
import statsmodels.api as sm

# Add a constant to the independent variable (X_train) for OLS regression
X_train_const = sm.add_constant(X_train)

# Initialize and train the OLS model
model_sm = sm.OLS(y_train, X_train_const).fit()

# Print the model summary, which includes the regression parameters
print(model_sm.summary())
```


## Predict on Test Data


Use the trained `statsmodels` OLS model to make predictions on the test set features.

## Predict on Test Data


Use the trained `statsmodels` OLS model to make predictions on the test set features.


```python
import statsmodels.api as sm

# Add a constant to the independent variable (X_test) for OLS prediction
X_test_const = sm.add_constant(X_test)

# Make predictions on the test data using the statsmodels OLS model
y_pred_sm = model_sm.predict(X_test_const)

print("Predictions on test data using statsmodels OLS model complete.")
```

## Visualize Model Predictions


Create a plot showing the training data points (normal color) and the predicted values for the test data (red color), along with the regression line.


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 7))

# Plot training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# Plot actual test data
plt.scatter(X_test, y_test, color='green', label='Actual Test Data')

# Plot predicted test data
plt.scatter(X_test, y_pred_sm, color='red', marker='x', s=100, label='Predicted Test Data')

# Get coefficients for the regression line
intercept = model_sm.params['const']
slope = model_sm.params['Daily walking (kms)']

# Create a range of x-values for the regression line based on the full dataset's 'Daily walking (kms)' range
x_range = np.linspace(df['Daily walking (kms)'].min(), df['Daily walking (kms)'].max(), 100)

# Calculate corresponding y-values for the regression line
y_regression_line = intercept + slope * x_range

# Plot the regression line
plt.plot(x_range, y_regression_line, color='orange', linestyle='--', label='Regression Line')

plt.xlabel('Daily walking (kms)')
plt.ylabel('Weight loss (kgs)')
plt.title('Linear Regression Model: Predictions vs. Actuals')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Evaluate the model using R-squared and RMSE
r2 = r2_score(y_test, y_pred_sm)
mse = mean_squared_error(y_test, y_pred_sm)
rmse = np.sqrt(mse)

print(f"R-squared: {r2:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
