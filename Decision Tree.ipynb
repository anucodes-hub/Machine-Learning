```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
my_file=pd.read_csv('/content/User_Data (3).csv');
```

```python
display(my_file)
```

```python
feature_col=['Age','EstimatedSalary'] #selecting features
x=my_file.iloc[:,[2,3]].values
y=my_file.iloc[:,4].values
```

```python
#split data
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
display(xtrain,xtest,ytrain,ytest)
```

```python
#scaling data-to remove effect of large values in training of model
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
xtrain=sc_X.fit_transform(xtrain)
xtest=sc_X.transform(xtest)
```

```python
print(xtrain)
```

```python
print(xtest)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier=classifier.fit(xtrain,ytrain)
```

```python
ypred=classifier.predict(xtest)
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(ytest,ypred))
cmf=confusion_matrix(ytest,ypred)
print(cmf)
```

```python
from sklearn.tree import export_graphviz
!pip install six
from six import StringIO
```

```python
from IPython.display import Image
import pydotplus
data = StringIO()
export_graphviz(classifier, out_file=data,
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_col,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(data.getvalue())
Image(graph.create_png())
```

```python
#decide optimum depth of tree depending on accuracy \
classifier_1=DecisionTreeClassifier(criterion='entropy',max_depth=5)
classifier_1=classifier_1.fit(xtrain,ytrain)
```

```python

ypred_1=classifier_1.predict(xtest)
print("Accuracy now:",accuracy_score(ytest,ypred_1))
```

```python
data = StringIO()
export_graphviz(classifier_1, out_file=data,
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_col,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(data.getvalue())
Image(graph.create_png())
```
