Importing Matplotlib library

```python
import matplotlib.pyplot as plt
```

To plot scatter-plot of Day vs Tip

```python
plt.scatter(my_file1['day'],my_file1['tip'],c=my_file1['size'],s=my_file1['total_bill'])

#c- color
#s- size

#color is according to size

#adding title to plot
plt.title('Scatter Plot: Day vs Tip')

#setting x and y label
plt.xlabel('Day')
plt.ylabel('Tip')

#adds color grading
plt.colorbar()

#show plot
plt.show()
```

**Observation:**

1.   On saturday maximum tip was given
2.   Color is according to size of dot which is according to total people
3. When Less no. of people,they pay lesser tip
