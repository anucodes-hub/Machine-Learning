Evaluation

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```


Build and evaluate a Gaussian Naive Bayes classifier to predict student placement status ('status') using the features 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', and 'mba_p'. After training and evaluating the model, visualize the relationships between these 5 features using a scatter plot matrix, colored by 'status'.

Select 5 Features: columns: 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', and 'mba_p'. The target variable (y) will remain 'status'.


```python
x = df[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']].values
print("Shape of x:", x.shape)
print("First 5 rows of x:\n", x[:5])
```

split the data into training and testing sets using `train_test_split` to prepare for model training.



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)
print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print("Gaussian Naive Bayes classifier retrained and predictions made.")
```

```python
from sklearn import metrics
print("Gaussian Naive Bayes accuracy in %",metrics.accuracy_score(y_test,y_pred)*100)
```
